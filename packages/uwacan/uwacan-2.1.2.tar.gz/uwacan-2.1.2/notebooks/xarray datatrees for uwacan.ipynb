```python
import xarray as xr
import numpy as np

class Accessor:
    def __init__(self, xr_obj):
        print('Attaching accessor')
        self._xr_obj = xr_obj

    @property
    def fooish(self):
        return 'fooish'

    def __call__(self, x):
        print('calling accessor with', x)

xr.register_dataarray_accessor('foo')(Accessor)

array = xr.DataArray(np.zeros(5))

array.foo(5)
```

```python
from datatree import DataTree
from xarray import DataArray, Dataset
import numpy as np
```

```python
channel_a = Dataset(
    {'power': (('frequency', 'time'), np.random.uniform(size=(15, 20)))},
    # coords={'time': np.arange(20), 'frequency': np.arange(30)},
    coords={'time': np.arange(20)},
    # dims=('frequency', 'time'),
    attrs={'hydrophone_depth': 10}
)
channel_b = Dataset(
    {'power': (('frequency', 'time'), np.random.uniform(size=(30, 20)))},
    # coords={'time': np.arange(20), 'frequency': np.arange(30)},
    coords={'time': np.arange(20)},
    # dims=('frequency', 'time'),
    attrs={'hydrophone_depth': 20}
)
# segment_1 = DataTree.from_dict({'channel_a': channel_a, 'channel_b': channel_b})
# segment_1 = DataTree(children={'channel_a': channel_a, 'channel_b': channel_b})
```

```python
channel_a_tree = DataTree(name='channel_a', data=channel_a)
channel_b_tree = DataTree(name='channel_b', data=channel_b)
```

```python
segment_1 = DataTree(name='segment_1', children={'channel_a': channel_a_tree, 'channel_b': channel_b_tree})
```

```python
segment_1 = DataTree(name='segment_1', data=Dataset(coords={'frequency': np.arange(30)}, attrs={'segment_anglee': 15}))
channel_a_tree.parent = segment_1
channel_b_tree.parent = segment_1
```
